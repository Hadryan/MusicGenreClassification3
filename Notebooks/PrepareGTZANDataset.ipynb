```python
import os
import pandas as pd
import numpy as np
import shutil
from tqdm import tqdm
from pydub import AudioSegment
```

```python
out_fld = '../data/GTZAN/data/'
```

```python
genres_map = {
 'pop': 0,
 'metal': 1,
 'disco': 2,
 'blues': 3,
 'reggae': 4,
 'classical': 5,
 'rock': 6,
 'hiphop': 7,
 'country': 8,
 'jazz': 9
}
```

```python
labels = []
for genre in tqdm(os.listdir('../data/GTZAN/genres/')):
    genre_path = '../data/GTZAN/genres/' + genre
    if os.path.isdir(genre_path):
        for file_name in os.listdir(genre_path):
            if '.DS_Store' in file_name:
                continue
            file_path = os.path.join(genre_path, file_name)
            elements = file_name.split('.')
            genre_id = genres_map.get(elements[0])

            track_id = "{:04d}".format(int(elements[1]))

            audio = AudioSegment.from_file(file_path, "au")
            audio.export(f"{out_fld}{genre_id}{track_id}.mp3", format="mp3")  

            labels.append([f'{genre_id}{track_id}', genre])
labels_df = pd.DataFrame(labels, columns=['track_id', 'genre'])
```

```python
labels_df.index.sort_values()
```

```python
labels_df.to_csv('../data/GTZAN/labels.csv')
```
